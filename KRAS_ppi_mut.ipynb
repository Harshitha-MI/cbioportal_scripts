```python
import pandas as pd
```

```python
string_ppi = pd.read_csv('/content/drive/MyDrive/Dr. V/string_interactions.tsv', sep = '\t')
string_ppi
```

```python
pan_mut = pd.read_excel('/content/drive/MyDrive/Dr. V/Pancreatic_KRAS_mutations.xlsx', sheet_name = 'Pan_KRAS_miss_mut - pan_mut',
                        header= 1)
pan_mut
```

```python
pan_main = pd.read_csv('/content/drive/MyDrive/Dr. V/Pancreatic Cancer_mutation_data.txt', sep = '\t')
pan_main.head()
```

Creat a list with all the protein-protein interactions of KRAS from the string_pathway file

```python
ppi = string_ppi.loc[string_ppi['#node1']== 'KRAS', 'node2'].tolist()
ppi += string_ppi.loc[string_ppi['node2']== 'KRAS', '#node1'].tolist()
ppi
```

Create a dataframe with patient details that have mutations in the genes upstream and downstream of KRAS.

```python
filtered_df = pan_main[(pan_main['PATIENT_ID'].isin(pan_mut['PATIENT_ID'])) &
                      (pan_main['Hugo_Symbol'].isin(ppi))]
filtered_df = filtered_df[['PATIENT_ID', 'Hugo_Symbol', 'Consequence','HGVSp']]
filtered_df.columns = ['PATIENT_ID', 'Mut_gene2', 'Consequence_g2', 'AA_change_g2']
filtered_df.shape
```

Create a merged dataframe with patient id's that have upstream and downstream, with respect to KRAS, gene mutations

```python
merged_df = pd.merge(pan_mut, filtered_df, on='PATIENT_ID', how='inner')
merged_df.shape
#pan_mut.shape
```

```python
merged_df.to_csv('/content/drive/MyDrive/Dr. V/KRAS_ppi_mut.txt', sep = '\t', index = False)
```
