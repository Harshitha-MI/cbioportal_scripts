Import data clinical sample and data mutations file

```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Dr. V/MSK-CHORD/Master_file_MSK-CHORD.csv', sep=',')
#print (data['CANCER_TYPE'].value_counts())
```

Filtering required columns from data_clinical

```python

select_col_filtered = ['SAMPLE_ID', 'PATIENT_ID','CANCER_TYPE','CANCER_TYPE_DETAILED','Hugo_Symbol','Consequence','Reference_Allele',
              'Tumor_Seq_Allele1','Tumor_Seq_Allele2','Tumor_Sample_Barcode','HGVSp']
data_filtered = data[select_col_filtered]
print(data_filtered.columns)
```

Types of cancers in MSK CHORD dataset

```python
cancer_type = data_filtered['CANCER_TYPE'].unique()
cancer_type = [str(cancer).replace('/', '_') for cancer in cancer_type]
#change des_cancer_type according to the cancer type of interest
des_cancer_type = ['Non-Small Cell Lung Cancer', 'Breast Cancer', 'Colorectal Cancer', 'Prostate Cancer', 'Pancreatic Cancer']
print(des_cancer_type)
```

For loop to generate cindividual cancer mutation files

```python
for cancer in des_cancer_type:
  cancer_data = data_filtered[data_filtered['CANCER_TYPE'] == cancer]
  file_name = f"{cancer.replace(' ', '_')}.csv"
  cancer_data.to_csv(f"/content/drive/MyDrive/Dr. V/MSK-CHORD/{file_name}", index=False)
  print(file_name, 'no. of samples' , len(cancer_data['SAMPLE_ID'].unique()))
  print(file_name , "no. of patient id's", len(cancer_data['PATIENT_ID'].unique()))
```
