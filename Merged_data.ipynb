Read in the data_clinical_sample file

```python
import pandas as pd

data_clinical = pd.read_csv('/content/data_clinical_sample.txt', sep='\t', skiprows = 4, header = 0)

#print (data_clinical.head())
```

Filtering the data based on cancer type and cancer type detailed

```python
filtered_data = data_clinical.loc[
    (data_clinical['CANCER_TYPE'] == 'Pancreatic Cancer') &
    (data_clinical['CANCER_TYPE_DETAILED'] == 'Pancreatic Adenocarcinoma')]

print(filtered_data.shape)
```

Extracting required columns

```python
#print(filtered_data.columns)

select_col_filtered = ['SAMPLE_ID', 'PATIENT_ID','CANCER_TYPE','CANCER_TYPE_DETAILED']

clinical = filtered_data[select_col_filtered]
print(clinical.columns)
```

Creating a list of sample id's

```python
print(filtered_data.columns)

sample_id = list(filtered_data['SAMPLE_ID'])
print(sample_id)
print(len(sample_id))
```

Reading in the data_mutation file

```python
data_mutation = pd.read_csv('/content/data_mutations.txt', sep='\t', header = 0)
#print (data_mutation.head())
print(data_mutation.columns)
```

Extracting required columns from data_mutation

```python
select_col = ['Hugo_Symbol','Consequence','Reference_Allele',
              'Tumor_Seq_Allele1','Tumor_Seq_Allele2','Tumor_Sample_Barcode','HGVSp']

data_mutation = data_mutation[select_col]
print(data_mutation.shape)
```

Extracting rows with pancreatic adenocarcinoma sample data using sample id list from filtered_data

```python
mutation = data_mutation[data_mutation['Tumor_Sample_Barcode'].isin(sample_id)]

print(mutation.shape)
```

Checking for multiple tumor sample barcode records

```python
barcode_counts = mutation['Tumor_Sample_Barcode'].value_counts()
print(barcode_counts)
```

Merging both the clinical and mutation data frames

```python
merged_data = pd.merge(mutation, clinical, left_on='Tumor_Sample_Barcode', right_on='SAMPLE_ID', how='left')
#print(merged_data.head)

print(merged_data.columns)
```

Reorder the columns in the merged_data dataframe

```python
reorder_col = ['PATIENT_ID','SAMPLE_ID','Tumor_Sample_Barcode','Hugo_Symbol','Consequence', 'HGVSp','Reference_Allele', 'Tumor_Seq_Allele1',
       'Tumor_Seq_Allele2', 'CANCER_TYPE', 'CANCER_TYPE_DETAILED']

merged_data = merged_data[reorder_col]

print(merged_data.head)
```

Saving the merged data frame

```python
merged_data.to_csv('/content/drive/MyDrive/merged_mutation.txt', sep='\t', index=False)
```
